```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(
    [[8,8,4],
     [7,9,5],
     [6,10,6],
     [5,12,7]],
    columns=['cgpa','profile_score','lpa']
)

df
```

```python
def initialize_parameters(layer_dims):

    np.random.seed(3)

    parameters = {}
    L = len(layer_dims)

    for l in range(1, L):

        parameters['W'+str(l)] = np.random.randn(layer_dims[l], layer_dims[l-1]) * 0.1

        parameters['b'+str(l)] = np.zeros((layer_dims[l],1))

    return parameters
```

```python
def linear_forward(A_prev, W, b):

    Z = np.dot(W, A_prev) + b

    return Z
```

```python
def L_layer_forward(X, parameters):

    A = X
    L = len(parameters) // 2

    caches = []

    for l in range(1, L+1):

        A_prev = A

        W = parameters['W'+str(l)]
        b = parameters['b'+str(l)]

        Z = linear_forward(A_prev, W, b)

        caches.append((A_prev, W, b))

        A = Z      

    return A, caches
```

```python
def compute_loss(y, y_hat):

    loss = np.square(y - y_hat)

    return loss
```

```python
def backward_propagation(y, y_hat, caches):

    grads = {}

    L = len(caches)

    
    dA = -2 * (y - y_hat)

    for l in reversed(range(L)):

        A_prev, W, b = caches[l]

        dW = np.dot(dA, A_prev.T)

        db = dA

        dA = np.dot(W.T, dA)

        grads["dW"+str(l+1)] = dW
        grads["db"+str(l+1)] = db

    return grads
```

```python
def update_parameters(parameters, grads, learning_rate):

    L = len(parameters) // 2

    for l in range(1, L+1):

        parameters["W"+str(l)] = parameters["W"+str(l)] - learning_rate * grads["dW"+str(l)]

        parameters["b"+str(l)] = parameters["b"+str(l)] - learning_rate * grads["db"+str(l)]

    return parameters
```

```python
X_train = df[['cgpa','profile_score']].values
Y_train = df[['lpa']].values

X_train = X_train.T
Y_train = Y_train.T

parameters = initialize_parameters([2,2,1])

epochs = 1000
learning_rate = 0.001

for i in range(epochs):

    total_loss = 0

    for j in range(X_train.shape[1]):

        X = X_train[:,j].reshape(2,1)
        y = Y_train[:,j].reshape(1,1)

        
        y_hat, caches = L_layer_forward(X, parameters)

        
        loss = compute_loss(y, y_hat)

        total_loss += loss

    
        grads = backward_propagation(y, y_hat, caches)

        
        parameters = update_parameters(parameters, grads, learning_rate)

    if i % 100 == 0:

        print("Epoch :", i, "Loss :", total_loss[0][0]/4)
```

```python
test = np.array([[9],[9]])

prediction, _ = L_layer_forward(test, parameters)

print("Predicted LPA :", prediction[0][0])
```
