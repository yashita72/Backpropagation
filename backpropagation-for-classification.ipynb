```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(
    [[8,8,1],
     [7,9,1],
     [6,10,0],
     [5,5,0]],
    columns=['cgpa','profile_score','placed']
)

df
```

```python
def initialize_parameters(layer_dims):

    np.random.seed(3)

    parameters = {}

    L = len(layer_dims)

    for l in range(1, L):

        parameters['W'+str(l)] = np.random.randn(layer_dims[l], layer_dims[l-1]) * 0.1

        parameters['b'+str(l)] = np.zeros((layer_dims[l],1))

    return parameters
```

```python
def sigmoid(Z):

    A = 1 / (1 + np.exp(-Z))

    return A
```

```python
def linear_forward(A_prev, W, b):

    Z = np.dot(W, A_prev) + b

    A = sigmoid(Z)

    return A
```

```python
def L_layer_forward(X, parameters):

    A = X

    L = len(parameters) // 2

    for l in range(1, L+1):

        A_prev = A

        W = parameters['W'+str(l)]

        b = parameters['b'+str(l)]

        A = linear_forward(A_prev, W, b)

    return A, A_prev
```

```python
def compute_loss(y, y_hat):

    loss = -y*np.log(y_hat) - (1-y)*np.log(1-y_hat)

    return loss
```

```python
def update_parameters(parameters, y, y_hat, A1, X):

    lr = 0.001

    parameters['W2'][0][0] = parameters['W2'][0][0] + (
        lr * (y - y_hat) * A1[0][0]
    )

    parameters['W2'][0][1] = parameters['W2'][0][1] + (
        lr * (y - y_hat) * A1[1][0]
    )

    parameters['b2'][0][0] = parameters['b2'][0][0] + (
        lr * (y - y_hat)
    )

    parameters['W1'][0][0] = parameters['W1'][0][0] + (
        lr * (y - y_hat)
        * parameters['W2'][0][0]
        * A1[0][0]
        * (1 - A1[0][0])
        * X[0][0]
    )

    parameters['W1'][0][1] = parameters['W1'][0][1] + (
        lr * (y - y_hat)
        * parameters['W2'][0][0]
        * A1[0][0]
        * (1 - A1[0][0])
        * X[1][0]
    )

    parameters['b1'][0][0] = parameters['b1'][0][0] + (
        lr * (y - y_hat)
        * parameters['W2'][0][0]
        * A1[0][0]
        * (1 - A1[0][0])
    )

    parameters['W1'][1][0] = parameters['W1'][1][0] + (
        lr * (y - y_hat)
        * parameters['W2'][0][1]
        * A1[1][0]
        * (1 - A1[1][0])
        * X[0][0]
    )

    parameters['W1'][1][1] = parameters['W1'][1][1] + (
        lr * (y - y_hat)
        * parameters['W2'][0][1]
        * A1[1][0]
        * (1 - A1[1][0])
        * X[1][0]
    )

    parameters['b1'][1][0] = parameters['b1'][1][0] + (
        lr * (y - y_hat)
        * parameters['W2'][0][1]
        * A1[1][0]
        * (1 - A1[1][0])
    )
```

```python
parameters = initialize_parameters([2,2,1])

epochs = 50

for i in range(epochs):

    Loss = []

    for j in range(df.shape[0]):

        X = df[['cgpa','profile_score']].values[j].reshape(2,1)

        y = df[['placed']].values[j][0]

        y_hat, A1 = L_layer_forward(X, parameters)

        y_hat = y_hat[0][0]


        loss = compute_loss(y, y_hat)

        Loss.append(loss)

        
        update_parameters(parameters, y, y_hat, A1, X)

    print("Epoch -", i+1,
          "Loss -", np.array(Loss).mean())
```

```python
X_test = np.array([[8],[9]])

y_pred, _ = L_layer_forward(X_test, parameters)

print(y_pred)
```

```python
if y_pred >= 0.5:
    print("Placed")
else:
    print("Not Placed")
```
